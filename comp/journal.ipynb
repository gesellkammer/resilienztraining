```python
def alldifftones(notesA, notesB):
    for ch0, ch1 in product(notesA, notesB):
        ch = ch0+ch1
        d = ch.difftones()[0]
        if not (n2m("4C") <= d.midi <= n2m("4G")):
            continue
        print(ch0.label, ch1.label, "   ", ch0[0].name + " + " + ch1[0].name)
        display(ch + ch.difftones())
```
